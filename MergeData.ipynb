```python
import pandas as pd
data=pd.read_csv("Data.csv")
data.head()
```

```python
data["City"]=data['City'].str.replace(' city.*' , '')
data["State"]=data['State'].str.strip()
data.head()
```

```python
data2=pd.read_csv("Data_Obesity.csv")
data2.head()
```

```python
Merge = pd.merge(data, data2, how='inner', on=['City','State', 'Year', 'Gender'])
Merge
```

```python
Merge.to_csv("CompleteData.csv",encoding="utf-8", index=False)
```
