```python
import pandas as pd
import gmaps
import json
import requests
import numpy as np

from config import geo_key
from config import owm_key

gmaps.configure(api_key=geo_key)
```

```python
city_data = pd.read_csv('DataAvrgFinal.csv')
city_data.head()
```

```python
some_city_data = city_data[city_data["AvrgCommute"]>35]
some_city_data.head()
```

```python
places = some_city_data["City"]
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{url}appid={owm_key}&units={units}&q="


lats = []
lngs = []
```

```python
for cities in places:
    response = requests.get(query_url + cities).json()
    if '404' in response.values():
        print(f"error 404")
        lats.append("NaN")
        lngs.append("NaN")
    else:
        print(f"Getting coodinates for city of {cities}")
    
        lats.append(response['coord']['lat'])
        lngs.append(response['coord']['lon'])
```

```python
lats_se = pd.Series(lats)
some_city_data['Lat'] = lats_se.values
#some_city_data.insert(loc=6, column='Latitude', value=lats_se)

lngs_se = pd.Series(lngs)
some_city_data['Lng'] = lngs_se.values
```

```python
locations = data_2013[["Lat", "Lng"]]

commute = data_2013["AvrgCommute"].astype(float)

fig_2013 = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=commute, 
                                 dissipating=False, max_intensity=60,
                                 point_radius=1)


# Add layer
fig_2013.add_layer(heat_layer)

# Display figure
fig_2013
```
